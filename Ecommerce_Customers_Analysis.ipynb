```python
emails= ecommerce['Email']
gmail= []
yahoo= []
hotmail= []
com= []
net= []
biz= []
org= []
info= []
others= []

for mail in emails:
    if '@gmail.com' in mail:
        gmail.append(mail)
    elif '@yahoo.com' in mail:
        yahoo.append(mail)
    elif '@hotmail.com' in mail:
        hotmail.append(mail)
    elif '.com' in mail:
        com.append(mail)
    elif '.net' in mail:
        net.append(mail)
    elif '.biz' in mail:
        biz.append(mail)
    elif '.org' in mail:
        org.append(mail)
    elif '.info' in mail:
        info.append(mail)
    else:
        others.append(mail)
```

```python
print(f'''
Numbers of gmail => {len(gmail)}
Numbers of yahoomail => {len(yahoo)}
Numbers of hotmail => {len(hotmail)}
Numbers of other Emails with .com => {len(com)}
Numbers of other Emails with .net => {len(net)}
Numbers of other Emails with .biz => {len(biz)}
Numbers of other Emails with .org => {len(org)}
Numbers of other Emails with .info => {len(info)}
''')
```
